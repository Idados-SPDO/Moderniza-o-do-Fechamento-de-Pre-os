```python
import pandas as pd
```

```python
df_planilhao = pd.read_excel(r"W:\\Otimização e Automatização\\5. Projetos\\7. Sistema de Fechamento de Preços\\Modernização de Fechamento de Preço\\Ferramenta_outlier - Reestruturação\data\\ALIVAR_30.10.24.xlsx", header=7)
df_planilhao
```

```python

validador = False

df_new = pd.DataFrame(columns=['Cód. Item Elementar', 'Desc. Item Elementar', 'Insinf/Cd Bases', 'Desc. Insinf', 'Sinônimo', 'Marca', 'Emb/Qtd', 
                               'Status Insinf', 'Cód Inf', 'Desc. Inf', 'Status Inf', 'Period', 'Preço Atual', 'Fator', 'Operador', 'Data Atual', 
                               'Preço Anterior', 'Preço Ref Anterior', 'Data Anterior', 'Tipo Preço', 'Região do preço', 'Reg. Ret.']) 

for i, j in df_planilhao.iterrows():
    if 'Analisado' in str(j['Nome']):
        validador = False 
    if 'Analisado' not in str(j['Nome']):
        validador = True
    if validador == False and isinstance(j['Insumo'], int):
        elementar = j['Insumo']
        descricao = j['Descrição']
        ref_anterior = j['$ Atu']
    if validador == True:
        new_row = [elementar, descricao, j['Ins.Inf.'], j['Descrição'], '-', j['Marca'], j['Embalagem'], j['St'],
                   j['Inform'], j['Nome'], j['St'], j['Period.'], j['Preço atu'], j['Fator atu'], j['Op atu'],
                   j['Dt.Colet.'], j['Preço ant'], ref_anterior, j['Dt.colet ant'], j['Tp Preço'], j['Unnamed: 11'], '-']
        df_new.loc[len(df_new)] = new_row
    

df_new = df_new.dropna(subset=['Insinf/Cd Bases'])
df_new = df_new[df_new['Data Anterior'] != ' ']

df_new
```
