```python
import pandas as pd
```

```python
df_planilhao = pd.read_excel(r"W:\\Otimização e Automatização\\5. Projetos\\7. Sistema de Fechamento de Preços\\Modernização de Fechamento de Preço\\Ferramenta_outlier - Reestruturação\data\\ALIATA_30.10.24.xlsx", header=7)
df_planilhao
```

```python

validador = False
df_new = pd.DataFrame(columns=['Cód. Item Elementar', 'Desc. Item Elementar', 'Insinf/Cd Bases', 'Desc. Insinf', 'Sinônimo', 'Marca', 'Emb/Qtd', 
                               'Status Insinf', 'Cód Inf', 'Desc. Inf', 'Status Inf', 'Period', 'Preço Atual', 'Fator', 'Operador', 'Data Atual', 
                               'Preço Anterior', 'Preço Ref Anterior', 'Data Anterior', 'Tipo Preço', 'Região do preço', 'Reg. Ret.']) 

for i, j in df_planilhao.iterrows():
    if validador == False and isinstance(j['Elementar'], int):
        elementar = j['Elementar']
        descricao = j['Descrição']
        ref_anterior = j['Preço ant']
    if validador == True:
        new_row = [elementar, descricao, j['Unnamed: 6'], j['Descrição'], '-', j['Unnamed: 2'], j['Medida'], j['Unnamed: 7'],
                   j['Unnamed: 4'], j['Unnamed: 5'], j['Unnamed: 7'], j['Busca'], j['Cota perf.'], j['Usuário Aprovador'], j['Data Última Aprovação do Item'],
                   j['Cota util'], j['Unnamed: 28'], ref_anterior, j['Unnamed: 29'], j['Variação'], j['Nível'], '-']
        df_new.loc[len(df_new)] = new_row
    if 'Referência' in str(j['Elementar']):
        validador = False 
    if 'Insumo' in str(j['Elementar']):
        validador = True

df_new = df_new.dropna(subset=['Insinf/Cd Bases'])
df_new = df_new[df_new['Data Anterior'] != ' ']

df_new
```
